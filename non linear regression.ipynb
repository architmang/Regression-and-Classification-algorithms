```python
df=pd.read_csv("china_gdp.csv")
x_data,y_data=df.Year,df.Value
plt.plot(df.Year,df.Value)
```

```python
def sigmoid(x,beta1,beta2):
    y=1/(1+np.exp(-(-beta2+x)*(beta1)))
    return y
```

```python
# to find optimum  parameters lets first normalize our data
x_data=x_data/max(x_data)
y_data=y_data/max(y_data)
```

```python
from scipy.optimize import curve_fit
from matplotlib import style
a,b=curve_fit(sigmoid,x_data,y_data)
# beta1=a[0]
# beta2=a[1]
x=np.arange(1960,2015,1)
x=x/max(x) # normalize x
y=sigmoid(x,*a)
plt.plot(x,y,label="fit",color="r",marker="o")
plt.plot(x_data,y_data,label="actual",color="b",marker="*")
plt.legend(loc="best",fontsize=20)
plt.minorticks_on()
plt.xlabel("year",fontsize=20)
plt.ylabel("GDP",fontsize=20)
style.use("seaborn")
plt.show()
```
